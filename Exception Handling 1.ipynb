```python
#try and else
"""
The try statement is used to enclose a block of code that may raise an exception.
If an exception is raised within this block of code, control is transferred to the except statement.
The else statement is executed if no exceptions were raised in the try block. i.e., there is no error inside the try block
"""  
try: 
    logging.info('Division') 
    x = 1 / 1 
except: 
    print("Exception Handling")  
else: 
    print("No Exception")
```

```python
#finally
"""
The finally statement is used to specify code that must be executed regardless of whether an exception was raised or not.
"""  
try: 
    logging.info('Division') 
    x = 1 / 1 
except: 
    print("Exception Handling")  
else: 
    print("No Exception")
finally:
    print("Universal")
```

```python
#5. What are Custom Exceptions in python? Why do we need Custom Exceptions? Explain with an example.
```

```python
"""
Custom Exceptions are user-defined exceptions that can be raised and caught in a program. 
Custom exceptions are created by creating a new class that inherits from the built-in Exception class.
"""
#Need of Custom Exceptions
"""
• Improved error handling.
• Code reuse.
• Better organization.
```

```python
#Example
import logging
balance = 2000
class Insufficient(Exception): 
    def __init__ (self, error): 
        self.error=error 

def withdraw(amount): 
    logging.info("Withdraw") 
    if amount > balance: 
        logging.error("Insufficient") 
        raise Insufficient("Not enough amount.") 
    else: 
        logging.info("Successful") 
        print("Transaction Completed")
```

```python
try: 
    amount = int(input("Enter: ")) 
    withdraw(amount) 
except Insufficient as e: 
    print(e) 
```

```python
class password(Exception): 
    def __init__(self,error): 
        self.error = error 

def Incorrect(pswd): 
    logging.info("validating Password") 
    if len(pswd)!=8: 
        logging.error("Incorrect Password") 
        raise Incorrect("Something went wrong") 
    else: 
        logging.info("Password Checked!")
```

```python
try: 
    pswd = input("Enter password: ") 
    Incorrect(pswd) 
except password as e: 
    print(e)
```
