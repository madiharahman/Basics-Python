```python
#ArithmeticError
"""The ArithmeticError exception class serves as a base class for those built-in exceptions that are raised due to various arithmetic errors."""
#like OverflowError, ZeroDivisionError, FloatingPointError.
```

```python
#ZeroDivisionError 
"""This error occurs when an attempt is made to divide a number by zero.""" 
# Example 
import logging
try: 
    21/0 
    logging.info("Checking Error")
except ZeroDivisionError as e: 
    logging.warning("Handling Error {} ".format(e))
    print("Successful! Zero Division Error")
```

```python
#OverflowError 
"""This error occurs when an arithmetic operation produces a result that is too large""" 
import logging,math 
try:
    math.exp(10000)
    logging.info("Checking Error")
except OverflowError as e: 
    logging.warning("Handling Error {} ".format(e))
    print("Successful! Overflow Error")
```

```python
#4. Why LookupError class is used? Explain with an example KeyError and IndexError.
```

```python
"""
The LookupError exception in Python forms the base class for all exceptions that are raised when an index or a key is not found 
for a sequence or dictionary respectively.
We can use LookupError exception class to handle both IndexError and KeyError exception classes.
"""
```

```python
#Key Error
import logging
try:
    x = {1:'madi',2:'fazzi',3:'saifi'}
    logging.info("Checking Error")
    print(x[4])
except LookupError as e:
    logging.warning("Handling Error {}".format(e))
    print("Successful! Key Error.")
```

```python
#Index Error
import logging
try:
    x = ['madi','fazzi','saifi']
    logging.info("Checking Error")
    print(x[4])
except LookupError as e:
    logging.warning("Handling Error {}".format(e))
    print("Successful! Index Error,list index out of range.")
```
